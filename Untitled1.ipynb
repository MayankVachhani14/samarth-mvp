```python
import pandas as pd

# reload the raw data
raw = pd.read_csv("data/rainfall.csv")

# 1) keep only needed columns
df = raw.copy()

# rename for clarity
df.rename(columns={"District": "district"}, inplace=True)

# 2) select year columns (those that contain "ANNUAL AVERAGE RAINFALL")
year_cols = [c for c in df.columns if "ANNUAL" in c]

# melt (convert wide → long)
df_long = df.melt(id_vars=["district"], value_vars=year_cols,
                  var_name="year_col", value_name="rainfall_mm")

# 3) extract numeric year from the column name
df_long["year"] = df_long["year_col"].str.extract(r"(\d{4})").astype(int)

# 4) clean district names and rainfall numbers
df_long["district"] = df_long["district"].astype(str).str.strip().str.title()
df_long["rainfall_mm"] = pd.to_numeric(df_long["rainfall_mm"], errors="coerce")

# 5) drop missing / invalid
df_long = df_long.dropna(subset=["rainfall_mm"])
df_long = df_long[["district", "year", "rainfall_mm"]].sort_values(["district","year"]).reset_index(drop=True)

# 6) save to parquet
df_long.to_parquet("data/rainfall_clean.parquet", index=False)

print("✅ cleaned file saved as data/rainfall_clean.parquet")
df_long.head(10)
```
